```python
import xarray as xr
import numpy as np
from datetime import datetime
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from scipy.stats import gaussian_kde
import os

plot_dir = os.path.expanduser("~/fig/")
os.makedirs(plot_dir, exist_ok=True)

# Flag
ship = "gj"

# File paths
file1 = f"/lfs/home/ioloi/lfs_dir/project_UCH/data/d00/ITRI_penghu.d00_hist_{ship}.nc"
file2 = f"/lfs/home/ioloi/lfs_dir/project_UCH/data/d00/ITRI_penghu.d00_ssp_{ship}.nc"
file3 = f"/lfs/home/ioloi/lfs_dir/project_UCH/data/d03{ship}_hist/ITRI_penghu.d03_hist_{ship}.domain.nc"
file4 = f"/lfs/home/ioloi/lfs_dir/project_UCH/data/d03{ship}_ssp/ITRI_penghu.d03_ssp_{ship}.domain.nc"
file5 = f"/lfs/home/ioloi/lfs_dir/project_UCH/data/d00_30yr/ITRI_penghu.d00_hist_{ship}.nc"
file6 = f"/lfs/home/ioloi/lfs_dir/project_UCH/data/d00_30yr/ITRI_penghu.d00_ssp_{ship}.nc"
output_dir = f"/lfs/home/ioloi/lfs_dir/project_UCH/post_result/"

# Time ranges for file1, file2, file3, file4
start_datetime = datetime(2012, 12, 2, 3, 15)
end_datetime = datetime(2013, 12, 27, 2, 45)

variables = ['temp', 'speed', 'pH']

def min_max_scaling(predicted, reference):
    """
    Rescale 'predicted' data so that its minimum and maximum match
    the minimum and maximum of 'reference' data using min‐max scaling.
    
    Formula:
      corrected = ((predicted - min_pred) / (max_pred - min_pred)) *
                  (max_ref - min_ref) + min_ref
    """
    min_pred = np.nanmin(predicted)
    max_pred = np.nanmax(predicted)
    min_ref = np.nanmin(reference)
    max_ref = np.nanmax(reference)
    
    # Prevent division by zero
    if max_pred - min_pred == 0:
        return np.full_like(predicted, min_ref)
    
    corrected = (predicted - min_pred) / (max_pred - min_pred) * (max_ref - min_ref) + min_ref
    return corrected

def read_and_select(file_path, variable, start_time, end_time):
    ds = xr.open_dataset(file_path)
    selected_data = ds[variable].sel(time=slice(start_time, end_time))
    print(f"File: {file_path}")
    print(f"Variable: {variable} | Time range: {start_time} to {end_time}")
    print(f"Min value: {selected_data.min().values}, Max value: {selected_data.max().values}\n")
    return selected_data

for variable in variables:
    print(f"Processing variable: {variable}")
    
    # ----------------- Training Data Preparation -----------------
    # Hist group: file1 (predictor) and file3 (target)
    data_x_hist = read_and_select(file1, variable, start_datetime, end_datetime)
    data_y_hist = read_and_select(file3, variable, start_datetime, end_datetime)
    
    # SSP group: file2 (predictor) and file4 (target)
    data_x_ssp = read_and_select(file2, variable, start_datetime, end_datetime)
    data_y_ssp = read_and_select(file4, variable, start_datetime, end_datetime)
    
    # Reference datasets for min‐max scaling (long‐term reference)
    ds5 = xr.open_dataset(file5)
    ref_hist = ds5[variable].values.flatten()  # Reference for hist group
    ds6 = xr.open_dataset(file6)
    ref_ssp = ds6[variable].values.flatten()    # Reference for ssp group
    
    # Apply min‐max scaling correction BEFORE training
    x_hist_corr = min_max_scaling(data_x_hist.values, ref_hist)
    y_hist_corr = min_max_scaling(data_y_hist.values, ref_hist)
    x_ssp_corr = min_max_scaling(data_x_ssp.values, ref_ssp)
    y_ssp_corr = min_max_scaling(data_y_ssp.values, ref_ssp)
    
    # ----------------- Combine Corrected Data and Train Model -----------------
    X_all = np.concatenate([x_hist_corr.flatten(), x_ssp_corr.flatten()]).reshape(-1, 1)
    y_all = np.concatenate([y_hist_corr.flatten(), y_ssp_corr.flatten()])
    
    X_train, X_val, y_train, y_val = train_test_split(X_all, y_all, test_size=0.2, random_state=42)
    
    model = RandomForestRegressor(random_state=42)
    model.fit(X_train, y_train)
    
    y_pred = model.predict(X_val)
    rmse = np.sqrt(mean_squared_error(y_val, y_pred))
    correlation = np.corrcoef(y_val, y_pred)[0, 1]
    
    plt.figure(figsize=(8, 6))
    plt.scatter(y_val, y_pred, alpha=0.6, label="Predicted vs True")
    plt.plot([min(y_val), max(y_val)], [min(y_val), max(y_val)],
             color='red', linestyle='--', label='Perfect Fit')
    plt.xlabel("True Values")
    plt.ylabel("Predicted Values")
    plt.title(f"Validation Scatter Plot ({variable})\nRMSE: {rmse:.2f}, Correlation: {correlation:.2f}")
    plt.legend()
    plt.grid(True)
    scatter_path = os.path.join(plot_dir, f"scatter_validation_{variable}_{ship}.png")
    plt.savefig(scatter_path)
    plt.show()
    
    # ----------------- Prediction on File5 (Hist) -----------------
    ds5_pred = xr.open_dataset(file5)
    data_new_hist = ds5_pred[variable]
    print(f"File: {file5}")
    print(f"Variable: {variable} | Reading all timesteps for prediction (Hist)")
    print(f"Min value: {data_new_hist.min().values}, Max value: {data_new_hist.max().values}\n")
    
    if variable == "temp":
        data_new_hist = data_new_hist.where(data_new_hist >= 10, drop=True)
    
    X_new_hist = data_new_hist.values.flatten().reshape(-1, 1)
    y_new_pred_hist = model.predict(X_new_hist)
    
    predicted_dataset_hist = xr.Dataset({
        f"predicted_{variable}": ("time", y_new_pred_hist),
    }, coords={"time": data_new_hist.time.values})
    
    output_path_hist = f"{output_dir}/predicted_{variable}_hist_{ship}.nc"
    predicted_dataset_hist.to_netcdf(output_path_hist)
    print(f"Predictions for File5 ({variable}) saved to {output_path_hist}")
    
    # ----------------- PDF Overlay for File5 -----------------
    # PDF of the original File5 data (reference distribution)
    kde_original_hist = gaussian_kde(data_new_hist.values.flatten())
    x_range_original_hist = np.linspace(np.nanmin(data_new_hist.values.flatten()),
                                        np.nanmax(data_new_hist.values.flatten()), 1000)
    pdf_original_hist = kde_original_hist(x_range_original_hist)
    
    # PDF of model predictions on File5
    kde_pred_hist = gaussian_kde(y_new_pred_hist)
    x_range_pred_hist = np.linspace(np.nanmin(y_new_pred_hist), np.nanmax(y_new_pred_hist), 1000)
    pdf_pred_hist = kde_pred_hist(x_range_pred_hist)
    
    # PDF of the training target data BEFORE correction (from File3)
    kde_train_orig_hist = gaussian_kde(data_y_hist.values.flatten())
    x_range_train_orig_hist = np.linspace(np.nanmin(data_y_hist.values.flatten()),
                                          np.nanmax(data_y_hist.values.flatten()), 1000)
    pdf_train_orig_hist = kde_train_orig_hist(x_range_train_orig_hist)
    
    # PDF of the training target data AFTER min‐max scaling (y_hist_corr)
    kde_train_corr_hist = gaussian_kde(y_hist_corr.flatten())
    x_range_train_corr_hist = np.linspace(np.nanmin(y_hist_corr.flatten()),
                                          np.nanmax(y_hist_corr.flatten()), 1000)
    pdf_train_corr_hist = kde_train_corr_hist(x_range_train_corr_hist)
    
    plt.figure(figsize=(8, 6))
    plt.plot(x_range_original_hist, pdf_original_hist, label="Reference Data (File5)", color='orange')
    plt.plot(x_range_pred_hist, pdf_pred_hist, label="Predicted (Model Output)", color='blue', alpha=0.7)
    plt.plot(x_range_train_orig_hist, pdf_train_orig_hist, label="Training Target (Original)", color='green', linestyle='--')
    plt.plot(x_range_train_corr_hist, pdf_train_corr_hist, label="Training Target (Corrected)", color='purple', linestyle=':')
    plt.xlabel(variable.capitalize())
    plt.ylabel("Density")
    plt.title(f"PDF of {variable.capitalize()} for File5 (Hist)\nReference vs Predicted vs Training Data (Before & After Correction)")
    plt.legend()
    plt.grid(True)
    plt.show()
    
    # ----------------- Prediction on File6 (SSP) -----------------
    ds6_pred = xr.open_dataset(file6)
    data_new_ssp = ds6_pred[variable]
    print(f"File: {file6}")
    print(f"Variable: {variable} | Reading all timesteps for prediction (SSP)")
    print(f"Min value: {data_new_ssp.min().values}, Max value: {data_new_ssp.max().values}\n")
    
    if variable == "temp":
        data_new_ssp = data_new_ssp.where(data_new_ssp >= 10, drop=True)
    
    X_new_ssp = data_new_ssp.values.flatten().reshape(-1, 1)
    y_new_pred_ssp = model.predict(X_new_ssp)
    
    predicted_dataset_ssp = xr.Dataset({
        f"predicted_{variable}": ("time", y_new_pred_ssp),
    }, coords={"time": data_new_ssp.time.values})
    
    output_path_ssp = f"{output_dir}/predicted_{variable}_ssp_{ship}.nc"
    predicted_dataset_ssp.to_netcdf(output_path_ssp)
    print(f"Predictions for File6 ({variable}) saved to {output_path_ssp}")
    
    # ----------------- PDF Overlay for File6 -----------------
    # PDF of the original File6 data (reference distribution)
    kde_original_ssp = gaussian_kde(data_new_ssp.values.flatten())
    x_range_original_ssp = np.linspace(np.nanmin(data_new_ssp.values.flatten()),
                                      np.nanmax(data_new_ssp.values.flatten()), 1000)
    pdf_original_ssp = kde_original_ssp(x_range_original_ssp)
    
    # PDF of model predictions on File6
    kde_pred_ssp = gaussian_kde(y_new_pred_ssp)
    x_range_pred_ssp = np.linspace(np.nanmin(y_new_pred_ssp), np.nanmax(y_new_pred_ssp), 1000)
    pdf_pred_ssp = kde_pred_ssp(x_range_pred_ssp)
    
    # PDF of the training target data BEFORE correction (from File4)
    kde_train_orig_ssp = gaussian_kde(data_y_ssp.values.flatten())
    x_range_train_orig_ssp = np.linspace(np.nanmin(data_y_ssp.values.flatten()),
                                         np.nanmax(data_y_ssp.values.flatten()), 1000)
    pdf_train_orig_ssp = kde_train_orig_ssp(x_range_train_orig_ssp)
    
    # PDF of the training target data AFTER min‐max scaling (y_ssp_corr)
    kde_train_corr_ssp = gaussian_kde(y_ssp_corr.flatten())
    x_range_train_corr_ssp = np.linspace(np.nanmin(y_ssp_corr.flatten()),
                                         np.nanmax(y_ssp_corr.flatten()), 1000)
    pdf_train_corr_ssp = kde_train_corr_ssp(x_range_train_corr_ssp)
    
    plt.figure(figsize=(8, 6))
    plt.plot(x_range_original_ssp, pdf_original_ssp, label="Reference Data (File6)", color='orange')
    plt.plot(x_range_pred_ssp, pdf_pred_ssp, label="Predicted (Model Output)", color='blue', alpha=0.7)
    plt.plot(x_range_train_orig_ssp, pdf_train_orig_ssp, label="Training Target (Original)", color='green', linestyle='--')
    plt.plot(x_range_train_corr_ssp, pdf_train_corr_ssp, label="Training Target (Corrected)", color='purple', linestyle=':')
    plt.xlabel(variable.capitalize())
    plt.ylabel("Density")
    plt.title(f"PDF of {variable.capitalize()} for File6 (SSP)\nReference vs Predicted vs Training Data (Before & After Correction)")
    plt.legend()
    plt.grid(True)
    plt.show()
```
